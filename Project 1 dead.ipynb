# Getting the Page need to scrape

```python
option = webdriver.ChromeOptions()  # this the option object of the webdriver
# headers = generate_user_agent(device_type="desktop", os=('win'))
```

```python
#option.add_argument('user_agent={}'.format(headers))
option.add_argument('--incognito --start-maximized')

browser = webdriver.Chrome(executable_path=r'D:\mypython\Weclouddata\TA\chromedriver.exe', options=option)
```

```python
url = 'https://www.accuweather.com/en/ca/mississauga/l5b/{}-weather/55071?year={}&view=list'

months = {1:'january',
          2:'february',
          3:'march',
          4:'april',
          5:'may',
          6:'june',
          7:'july',
          8:'august',
          9:'september',
          10:'october',
          11:'november',
          12:'december'}

year = range(2009,2010)

for y in year:
    for i in range(1,13):
              
        browser.get(url.format(months[i],y))
       
        date_s=[]
        high=[]
        low=[]
        precip=[]
        HVhigh=[]
        HVlow=[]
        snow=[]

        for i in range(1,32):
            try:
                d = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[1]/div/div/div[1]/p[2]').text
                date_s.append(d+'/'+str(y))
                h = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[1]/div/div/div[2]/span[1]').text
                high.append(h[:-1])
                l = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[1]/div/div/div[2]/span[2]').text
                low.append(l.split()[1][:-1])  
                p = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[1]/div/div/div[3]/p[2]').text
                precip.append(p.split()[0])
                hvh = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[2]/div/div[2]/p[1]').text
                HVhigh.append(hvh)
                hvl = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[2]/div/div[2]/p[2]').text
                HVlow.append(hvl.split()[3][:-1])
                s = browser.find_element_by_xpath(f'/html/body/div/div[5]/div/div[1]/div/div[2]/div[{i}]/div[2]/div/div[1]/p[3]').text
                if s=='':
                    snow.append('0')
                else:
                    snow.append(s.split()[1])
            except NoSuchElementException:
                snow.append('0')
            continue
        HV_high=[]
        HV_low=[]
        for hvh in HVhigh:
            try:
                HV_high.append(hvh.split()[3][:-1])
                
            except:
                pass
            continue
        for hvl in HVlow:
            try:
                HV_low.append(hvl.split()[3][:-1])
                
            except:
                pass
            continue    
        dict = {'date': date_s, 'high(째C)': high, 'low(째C)': low,'precip(mm)': precip, 'snow(mm)': snow, 'HV_high(째C)': HV_high,'HV_low(째C)':HV_low}

        df = pd.DataFrame(dict)

        df['date'] = pd.to_datetime(df['date'])

        df.set_index('date', inplace=True)

        df.to_csv(r'D:\mypython\Weclouddata\Project 1\{}_{}.csv'.format(y,i))
        
        browser.close()

        sleep(10+10*random.random()) # sleep(10) stops the notebook for 10 seconds. I use sleep here to wait for loading the page.
```
